```python
# Import Libraries.
import nltk
from nltk import word_tokenize
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from nltk.corpus import sentiwordnet as swn
from nltk.sentiment.vader import SentimentIntensityAnalyzer

import time
```

```python
annotations = pd.read_csv("../data/fb_annotations.csv")
annotations
```

```python
annot_nans = annotations[annotations.isna().any(axis=1)]
annot_nans
```

```python
# Run Vader sentiment analysis on each response in the annotations set.
start_time = time.time()
sia = SentimentIntensityAnalyzer()
Vader_sent = annotations["response_text"].apply(sia.polarity_scores)
Vader_sent = [d.get('compound') for d in Vader_sent]
annotations['Vader_sent'] = Vader_sent
print("run time:",time.time()-start_time)
annotations
```

```python
def convert_sentiment(x):
    if x == 'Negative':
        return -1
    elif x == 'Positive':
        return 1
    else:
        return 0
```

```python
def convert_Vader(x):
    if x < 0:
        return -1
    elif x == 0:
        return 0
    else:
        return 1
```

```python
annotations['Vader_sent_01'] = annotations['Vader_sent'].apply(convert_Vader)
annotations['sent_01'] = annotations['sentiment'].apply(convert_sentiment)
annotations
```

```python
V_comp = annotations['sent_01'] == annotations['Vader_sent_01']
V_comp.value_counts()
```
