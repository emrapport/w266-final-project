```python
# Run Vader sentiment analysis on each response.
start_time = time.time()
Vader_sent = responses["response_text"].apply(sia.polarity_scores)
Vader_sent = [d.get('compound') for d in Vader_sent]
responses['Vader_sent'] = Vader_sent
print("run time:",time.time()-start_time)
```
